```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "../Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(5)
```

## Player Count

* Display the total number of players


```python
total_number_players=len(purchase_data["SN"].unique())
total_number_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items=len(purchase_data["Item Name"].unique())
avg_price=purchase_data["Price"].mean()
no_purchase=purchase_data["Purchase ID"].count()
total_revenu=purchase_data["Price"].sum()
summary_df=pd.DataFrame({'Number of Unique Items':[unique_items],
         'Average Price':[avg_price],
         'Number of Purchases':[no_purchase],
         'Total Revenu':[total_revenu]
    
})
#summary_df = pd.DataFrame(summary,columns=['Number of Unique Items','Average Price', 'Number of Purchases','Total Revenu'])
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#drop any user duplicates and save in counts df
purchase_data_counts=purchase_data.drop_duplicates(subset=["SN"])
#get value counts and convert to DF
gender_df=pd.DataFrame(purchase_data_counts["Gender"].value_counts())
#remane column
gender_df=gender_df.rename(columns={"Gender":"Total Count"})
#add percentage column
gender_df["Percentage of Players"]=purchase_data_counts["Gender"].value_counts()/total_number_players*100
#change percentage format
gender_df["Percentage of Players"]=gender_df["Percentage of Players"].map("{:.2f}%".format)
gender_df

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_analysis=purchase_data.groupby("Gender")
analysis_df=pd.DataFrame(purchase_analysis["Price"].count())
#add avg purchase price column
analysis_df["Average Purchase Price"]=purchase_analysis["Price"].mean().map("${:.2f}".format)
#add Totals
analysis_df["Total Purchase Value"]=purchase_analysis["Price"].sum().map("${:.2f}".format)
#add avg totals
analysis_df["Average Total Purchase per Person"]=(purchase_analysis["Price"].sum()/purchase_analysis["Price"].count()).map("${:.2f}".format)
analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create the bins
bins=[0,9,14,19,24,29,34,39,100]
#name the bins
bins_df=purchase_data.drop_duplicates(subset=["SN"])
#bins_df.rename(columns={"Age":"Total Count"})

bin_name=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
bins_df["Age range"] = pd.cut(bins_df["Age"], bins, labels=bin_name, include_lowest=True)
bins_df=bins_df.groupby("Age range")
sum_df=pd.DataFrame(bins_df["Age"].count())
sum_df["Percentage Players"]=(bins_df["Age"].count()*100/total_number_players).map("{:.2f}%".format)
sum_df=sum_df.rename(columns={"Age":"Total Count"})
sum_df
```
